**Q8.**  Write a function *my_fun_plotter(f, x)*, where f is a lambda function and x is an array. The function should plot f evaluated at x. Remember to label the x- and y-axes.

```python
def my_fun_plotter(f, x):
    plt.plot(x, f(x))
    plt.axhline(color = 'r', linewidth = .7, linestyle = '--')
    plt.axvline(color = 'r', linewidth = .7, linestyle = '--')
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.axis('equal')
    plt.show()
    
my_fun_plotter(lambda x: np.sqrt(x) + np.exp(np.sin(x)), np.linspace(0, 2*np.pi, 100))
my_fun_plotter(lambda x: np.sin(x), np.linspace(0, 4*np.pi, 100))
```

**Q9.** Write a function with *my_poly_plotter(n, x)* that plots the polynomials $p_k(x) = x^k$ for k = 1, . . . , n. Make sure your plot has axis labels and a title.

```python
def my_poly_plotter(n, x):
    colors = plt.cm.viridis(np.linspace(0, 1, n+1))
    for i in range(1, n+1):
        plt.plot(x, x**i)
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.axis('equal')
    plt.title(f'Polynomial up to degree {n}')
    plt.show()
    
my_poly_plotter(5, np.linspace(-1, 1, 200))
```

**Q10.**  Assume you have three points at the corner of an equilateral triangle, $P1 = (0, 0),$ $P2 = (0.5, \frac{\sqrt{2}}{2}),$ and $P3 =(1,0)$. Generate another set of points $p_i = (x_i, y_i)$ such that $p_1 = (0,0)$ and $p_{i+1}$ is the midpoint between $p_i$ and $P_1$ with 33% probability, the midpoint between $p_i$ and $P_2$ with 33% probability, and the midpoint between $p_i$ and $P_3$ with 33% probability. Write a function *my_sierpinski(n)* that generates the points $p_i$ for i = 1, . . . , n. The function should make a plot of the points using blue dots (i.e., "b." as the third argument to the plt.plot function).

```python
def my_sierpinski(n):
    P1 = (0, 0)
    P2 = (0.5, (2**0.5)/2)
    P3 = (1, 0)
    p = P1
    points = [p]
    
    for i in range(n):
        rand = random.random()
        if rand < 1/3:
            p = ((p[0] + P1[0]) / 2, (p[1] + P1[1]) / 2)
        elif rand < 2/3:
            p = ((p[0] + P2[0]) / 2, (p[1] + P2[1]) / 2)
        else:
            p = ((p[0] + P3[0]) / 2, (p[1] + P3[1]) / 2)
        points.append(p)
    x, y = zip(*points)
    plt.plot(x, y, 'b.', markersize = 1)
    plt.title("Sierpinski Triangle")
    plt.show()
    
my_sierpinski(100)
my_sierpinski(10000)
```

**Q11.**  Assume you are generating a set of points $(x_i, y_i),$ where $x_1 = 0$ and $y_1 = 0.$ The points $(x_i, y_i)$ for $i = 2,...,n$ are generated according to the following probabilistic relationship:

With 1% probability: $x_i = 0, y_i = 0.16y_{i1};$


With 7% probability: $x_i = 0.2x_{i1}  0.26y_{i1}, y_i = 0.23x_{i1} + 0.22y_{i1} + 1.6;$


With 7% probability: $x_i = 0.15x_{i1} + 0.28y_{i1}, y_i = 0.26x_{i1} + 0.24y_{i1} + 0.44;$


With 85% probability: $x_i = 0.85x_{i1} + 0.04y_{i1}, y_i = 0.04x_{i1} + 0.85y_{i1} + 1.6.$


Write a function *my_fern(n)* that generates the points $(x_i, y_i)$ for $i = 1,...,n$ and plots them using blue dots. Also use *plt.axis("equal")* and *plt.axis("off")* to make the plot more attractive.

```python
def my_fern(n):
    p = (0, 0)
    P = [p]
    
    for i in range(n):
        rand = random.randint(1, 100)
        if rand < 2:
            p = (0, 0.16*p[1])
        elif rand in range(2, 9):
            p = (0.2*p[0] - 0.26*p[1], 0.23*p[0] + 0.22*p[1] + 1.6)
        elif rand in range(9, 16):
            p = (-0.15*p[0] + 0.28*p[1], 0.26*p[0] + 0.24*p[1] + 0.44)
        else:
            p = (0.85*p[0] + 0.04*p[1], -0.04*p[0] + 0.85*p[1] + 1.6)
        P.append(p)
    x, y = zip(*P)
    plt.plot(x, y, 'b.', markersize = 1)
    plt.axis("equal")
    plt.axis("off")
    plt.title(f"My Fern with {n} iterations")
    plt.show()
    
my_fern(100)
my_fern(10000)
```

**Q12.** Write a function *my_parametric_plotter (x, y, t)*, where x and y are function objects x(t) and y(t), respectively, and t is a one-dimensional array. The function *my_parametric_plotter* should produce the curve (x(t), y(t), t) in a 3D plot. Be sure to give your plot a title and label your axes.

```python
def my_parametric_plotter(x, y, t):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x(t), y(t), t)
    ax.set_xlabel('x-axis')
    ax.set_ylabel('y-axis')
    ax.set_zlabel('t-axis')
    ax.grid(False)
    plt.title('My Parametric Plot')
    plt.show()
    
f = lambda t: np.sin(t)
g = lambda t: t**2
my_parametric_plotter(f, g, np.linspace(0, 6*np.pi, 100))
```
